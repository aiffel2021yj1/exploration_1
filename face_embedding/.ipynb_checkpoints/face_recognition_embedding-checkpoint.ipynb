### 2. 얼굴 영역 자르기

```python
image_path = os.getenv('HOME')+'/aiffel/exploration_1/face_embedding/image/hyori_lee_0.jpg'
image = face_recognition.load_image_file(image_path)
face_locations = face_recognition.face_locations(image)

print(face_locations)  # 이미지에서 얼굴 영역의 좌표를 출력합니다.

a, b, c, d = face_locations[0]
cropped_face = image[a:c,d:b,:]

plt.imshow(cropped_face)   # 이미지에서 얼굴영역만 잘라낸 cropped_face를 그려 봅니다.
```

##### 이미지 파일 경로를 파라미터로 넘기면 얼굴 영역만 잘라서 뿌려주는 함수

```python
def get_cropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_face = image[a:c,d:b,:]
    
    return cropped_face
```

### 3. 얼굴 영역 임베딩 추출하기

##### embedding function

```python
image_file = os.path.join(dir_path, 'hyori_lee_0.jpg')
face = get_cropped_face(image_file)

# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
embedding
```

##### store values a dictionary 

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_cropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0:  
        # 얼굴영역 face가 제대로 detect되지 않으면  len(embedding)==0인 경우가 발생하므로 
        # os.path.splitext(file)[0]에는 이미지파일명에서 확장자를 제거한 이름이 담깁니다. 
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
        
    return embedding_dict
```

```python
embedding_dict = get_face_embedding_dict(dir_path)
print(len(embedding_dict))
```

### 4. 비교하기
- 임베딩 벡터들 간의 거리를 비교해보고 가장 작은 경우 찾기

##### comparing the distance between two subjects

```python
def get_distance(name1, name2):
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)
```

```python
get_distance('hyori_lee_0', 'taylor_swift_0')
```

```python
get_distance('taylor_swift_0', 'taylor_swift_1')
```

* 닮은 꼴의 두 연예인의 사진보다 동일 연예인 인물 사진이 더 비슷하다 (당연하겠지만 동일인물도 연출법에 따라 달라보일 수 있기 때문에 궁금했다)

```python
def get_nearest_face(name, top=5):
    sort_key = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key(x[0]))
    
    for i in range(top+1):
        if i == 0:
            continue
        if sorted_faces[i]:
            print(f"top {i}: name({sorted_faces[i][0]}), distance({sort_key(sorted_faces[i][0])})")
```

```python
# name1과 name2의 거리를 비교하는 함수를 생성하되, name1은 미리 지정하고, name2는 호출시에 인자로 받도록 합니다.
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1

sort_key_func = get_sort_key_func('hyori_lee_0')   
```
